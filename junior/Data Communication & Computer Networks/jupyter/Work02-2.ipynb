```python
import urllib
from sigproc import Signal,start_notebook
start_notebook()

urllib.request.urlretrieve("https://www.cpe.ku.ac.th/~cpj/204325/q3-2/q3-2-49.wav","q3-2-49.wav")
s = Signal(wav_file="q3-2-49.wav")
s.plot_time(fig_options=dict(x_range=(0,100)))
```

```python
s.plot_freq(min_amplitude=0.01,fig_options=dict(height=200))
s.plot_phase(min_amplitude=0.01,fig_options=dict(height=200))
```

```python
from numpy import cos,pi,radians

def f(t):
    return (
        0.1*cos(2*pi*500*t+pi/2) + 0.3*cos(2*pi*700*t+pi) + 0.2*cos(2*pi*900*t+3*pi/2) 
    )

s2 = Signal(func=f)
s2.plot_time(fig_options=dict(x_range=(0,100)))
s.plot_time(fig_options=dict(x_range=(0,100)))
```
