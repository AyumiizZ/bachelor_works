```python
import urllib
from sigproc import Signal,start_notebook
bkp = start_notebook()

urllib.request.urlretrieve("https://www.cpe.ku.ac.th/~cpj/204325/q7-1/q7-1-55.wav","q7-1-55.wav")
s = Signal(wav_file="q7-1-55.wav")
s.plot_time()
s.plot_freq(fig_options=dict(x_range=(100,1200),height=200))
```

```python
ch2 = s.copy().filter((200,400))
ch2.plot_freq(fig_options=dict(x_range=(0,500),height=200))
ch2.plot_time(fig_options=dict(height=200))
```

```python
freq = 300
from numpy import sin,cos,pi
carrier_i = Signal(duration=ch2.duration,func=lambda t:sin(2*pi*freq*t))
carrier_q = Signal(duration=ch2.duration,func=lambda t:cos(2*pi*freq*t))
baseband_i = (ch2*carrier_i*2).filter(freq)
baseband_q = (ch2*carrier_q*2).filter(freq)
print("Baseband I")
baseband_i.plot_time(fig_options=dict(height=200))
print("Baseband Q")
baseband_q.plot_time(fig_options=dict(height=200))
```

```python
from numpy import sin,cos,pi
for freq in [300,600,900]:
    ch2 = s.copy().filter((freq-100,freq+100))
#     ch2.plot_freq(fig_options=dict(x_range=(freq-200,freq+200),height=200))
#     ch2.plot_time(fig_options=dict(height=200))
    
    carrier_i = Signal(duration=ch2.duration,func=lambda t:sin(2*pi*freq*t))
    carrier_q = Signal(duration=ch2.duration,func=lambda t:cos(2*pi*freq*t))
    baseband_i = (ch2*carrier_i*2).filter(freq)
    baseband_q = (ch2*carrier_q*2).filter(freq)
    print("Baseband I: " + str(freq))
    baseband_i.plot_time(fig_options=dict(height=200))
    print("Baseband Q: " + str(freq))
    baseband_q.plot_time(fig_options=dict(height=200))
```
