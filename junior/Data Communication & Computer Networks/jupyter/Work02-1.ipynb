```python
from numpy import sin,cos,pi
def sine1(t):
    return 0.7*cos(2*pi*800*t+pi/2)
def sine2(t):
    return sin(2*pi*1000*t)
s1 = Signal(func=sine2)
s1.plot_time(fig_options=dict(x_range=(0,5)))
```

```python
s1.plot_freq(min_amplitude=0.01,fig_options=dict(height=200))
s1.plot_phase(min_amplitude=0.01,fig_options=dict(height=200,x_range=(500,1500)))
```
