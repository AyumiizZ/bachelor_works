```python
import urllib
from sigproc import Signal,start_notebook
bkp = start_notebook()

urllib.request.urlretrieve("https://www.cpe.ku.ac.th/~cpj/204325/q6-3/q6-3-53.wav","q6-3-53.wav")
s = Signal(wav_file="q6-3-53.wav")
s.plot_time()
```

```python
from numpy import sin,cos,pi
duration = s.duration
ci = Signal(duration=s.duration,func=lambda t: sin(2*pi*550*t))
cq = Signal(duration=s.duration,func=lambda t: cos(2*pi*550*t))
```

```python
ci.plot_time_and_freq()
cq.plot_time_and_freq()
```

```python
(s*ci*2).plot_time_and_freq()
```

```python
i = (s*ci*2).filter(550)
q = (s*cq*2).filter(550)
i.plot_time_and_freq()
q.plot_time_and_freq()
```
